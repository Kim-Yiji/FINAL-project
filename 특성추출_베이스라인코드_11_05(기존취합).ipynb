# 특성 공학(Feature Engineering)

## 날짜 형식으로 변환

```python
train_tr["구매일시"] = pd.to_datetime(train_tr["구매일시"])
test_tr["구매일시"] = pd.to_datetime(test_tr["구매일시"])
```

## 새로 만든 feature와 병합할 고객ID로만 이루어진 데이터프레임 생성

```python
train_ft = train_target[["ID"]]
test_ft = submit[["ID"]]

train_ft.shape, test_ft.shape
```

## 구매일시를 이용한 특성생성

```python
agg_list = [
        # 컬럼명, 집계 방식
        ('구매횟수','nunique'),
        ('내점일수', lambda x: x.dt.date.nunique()),
        ('구매주기', lambda x: int( (x.max() - x.min()).days / x.dt.date.nunique()) ),
        ('주말방문비율', lambda x: np.mean(x.dt.weekday>4)),
        ('봄_구매비율', lambda x: np.mean(x.dt.month.isin([3,4,5]))),
        ('여름_구매비율', lambda x: np.mean(x.dt.month.isin([6,7,8]))),
        ('가을_구매비율', lambda x: np.mean(x.dt.month.isin([9,10,11]))),
        ('겨울_구매비율', lambda x: np.mean(x.dt.month.isin([1,2,12]))),
        ('주구매요일', lambda x: x.dt.weekday.mode()[0]),
        ('일별평균구매건수', lambda x:  x.count() / x.dt.date.nunique() ),
        ('거래개월수', lambda x: x.dt.date.astype(str).str[:-3].nunique() ),

        #추가 내용
        ('12시이전구매비율', lambda x: np.mean(x.dt.hour <= 12)),
        ('12시이후_18시이전구매비율', lambda x: np.mean((x.dt.hour > 12) & (x.dt.hour < 18))),
        ('18시이후구매비율', lambda x: np.mean(x.dt.hour >= 18)),
        ('주중방문비율', lambda x : np.mean(x.dt.weekday < 5)),
        ('주말방문비율', lambda x : np.mean(x.dt.weekday >= 5))
    ]

tmp = train_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

```python
#추가내용
train_ft['많은_구매주기'] = train_ft['구매주기'] > train_ft['구매주기'].median()
train_ft['많은_거래개월수'] = train_ft['거래개월수'] > train_ft['거래개월수'].median()
train_ft['구매주기_거래개월'] = train_ft['많은_거래개월수'] & train_ft['많은_구매주기']

test_ft['많은_구매주기'] = train_ft['구매주기'] > train_ft['구매주기'].median()
test_ft['많은_거래개월수'] = train_ft['거래개월수'] > train_ft['거래개월수'].median()
test_ft['구매주기_거래개월'] = train_ft['많은_거래개월수'] & train_ft['많은_구매주기']
```

## 지점을 이용한 특성생성

```python
agg_list = [
          ("방문지점수","nunique"),
          ('주구매지점', lambda x: x.mode()[0]),
    ]

tmp = train_tr.groupby('ID')["지점코드"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["지점코드"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

## 브랜드코드를 이용한 특성생성

```python
agg_list = [
             ('브랜드코드_nunique', 'nunique'),
             ]

tmp = train_tr.groupby('ID')["브랜드코드"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["브랜드코드"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

## 중분류를 이용한 특성생성

```python
agg_list = [
            ('중분류_nunique', 'nunique'),
            ('주구매_중분류', lambda x: x.mode()[0]),
            #추가내용
            ('중분류_아동_cnt', lambda x: x.str.contains('아동').sum())
            ]
tmp = train_tr.groupby('ID')["중분류"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left')
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["중분류"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

## 대분류를 이용한 특성생성

#⚓
여성,남성,생식품추가

```python
agg_list = [
            ('대분류_nunique', 'nunique'),
            ('주구매_대분류', lambda x: x.mode()[0]),
            #추가내용
            ('대분류_아동_cnt', lambda x: x.str.contains('아동').sum()),
            ('대분류_여성_cnt', lambda x: x.str.contains('여성').sum()),
            ('대분류_남성_cnt', lambda x: x.str.contains('남성').sum()),
            ('대분류_생식품_cnt', lambda x: x.str.contains('생식품').sum())
            ]

tmp = train_tr.groupby('ID')["대분류"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left')
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["대분류"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

## 구매가격을 이용한 특성생성

```python
agg_list = [
        ('총구매액','sum'),
        ('구매건수', 'count'),
        ('평균구매액', "mean"),
        ('최대구매액', 'max'),
        ('최소구매액',lambda x: x[x > 0].min() ) ,
        ('환불금액',lambda x: x[x < 0].sum() ) ,
        ('환불건수', lambda x: ( x < 0 ).sum() ),
        ('구매금액표준편차',lambda x: x[x>0].std() ),

        #추가내용
        ('구매금액왜도', lambda x: x[x > 0].skew()),
        ('구매금액첨도', lambda x: x[x > 0].kurtosis()),
        ('구매금액_1q', lambda x: x[x > 0].quantile(0.25)),
        ('구매금액_2q', lambda x: x[x > 0].quantile(0.5)),
        ('구매금액_3q', lambda x: x[x > 0].quantile(0.75))



    ]

tmp = train_tr.groupby('ID')["구매가격"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left')
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["구매가격"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

```python
#추가내용
train_ft["실제구매횟수"] = train_ft["구매건수"] - (train_ft["환불건수"]*2)
train_ft["환불율"] = train_ft["환불건수"] / (train_ft["구매건수"]-train_ft['환불건수'])
test_ft["실제구매횟수"] = test_ft["구매건수"] - (test_ft["환불건수"]*2)
test_ft["환불율"] = test_ft["환불건수"] / (test_ft["구매건수"]-test_ft['환불건수'])

train_ft['많은_구매액'] = (train_ft['최소구매액'] >= 10000) & (train_ft['최대구매액'] > train_ft['최대구매액'].median())
test_ft['많은_구매액'] = (train_ft['최소구매액'] >= 10000) & (train_ft['최대구매액'] > train_ft['최대구매액'].median())
```

```python
# 시간별 구매횟수 추가
agg_list = [
        ('12시이전구매횟수', lambda x: (x.dt.hour <= 12).sum()),
        ('12시이후_18시이전구매횟수', lambda x: ((x.dt.hour > 12) & (x.dt.hour < 18)).sum()),
        ('18시이후구매횟수', lambda x: (x.dt.hour >= 18).sum()),
        ('방문평균거래평균횟수', lambda x: x.nunique()/x.dt.date.nunique())

    ]

tmp = train_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
tmp = test_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
```

```python
# 브랜드별 구매3회이상 추가
brand_counts = train_tr.groupby('ID')['브랜드코드'].value_counts()
brand_counts_filtered = brand_counts[brand_counts >= 3].reset_index(name='count')
tmp = brand_counts_filtered["ID"]
tmp = tmp.drop_duplicates()
tmp=tmp.reset_index()
tmp['3회이상count'] = 1
tmp = tmp.drop(columns=['index'])
train_ft = train_ft.merge(tmp, how='left',on="ID")

brand_counts = test_tr.groupby('ID')['브랜드코드'].value_counts()
brand_counts_filtered = brand_counts[brand_counts >= 3].reset_index(name='count')
tmp = brand_counts_filtered["ID"]
tmp = tmp.drop_duplicates()
tmp=tmp.reset_index()
tmp['3회이상count'] = 1
tmp = tmp.drop(columns=['index'])
test_ft = test_ft.merge(tmp, how='left',on="ID")
```

```python
# 남녀_모두구매
train_ft["남_여모두구매"]=(train_ft["대분류_남성_cnt"]>0) & (train_ft["대분류_여성_cnt"]>0)
test_ft["남_여모두구매"]=(test_ft["대분류_남성_cnt"]>0) & (test_ft["대분류_여성_cnt"]>0)
# 대분류가 생식품인 가격의 합을 ID별로 그룹화
price_sum_by_id = train_tr[train_tr['대분류'] == '생식품'].groupby('ID')['구매가격'].sum().reset_index()
# Rename the column to "총_생식품_구매가격"
price_sum_by_id = price_sum_by_id.rename(columns={'구매가격': '총_생식품_구매가격'})
train_ft = train_ft.merge(price_sum_by_id,on="ID", how='left')
train_ft.loc[train_ft["대분류_생식품_cnt"] > 0, "평균_생식품_구매가격"] = train_ft["총_생식품_구매가격"] / train_ft["대분류_생식품_cnt"]
train_ft.loc[train_ft["대분류_생식품_cnt"] <= 0, "평균_생식품_구매가격"] = 0
#Repeat for test_ft
price_sum_by_id = test_tr[test_tr['대분류'] == '생식품'].groupby('ID')['구매가격'].sum().reset_index()
# Rename the column to "총_생식품_구매가격"
price_sum_by_id = price_sum_by_id.rename(columns={'구매가격': '총_생식품_구매가격'})
test_ft = test_ft.merge(price_sum_by_id,on="ID", how='left')
test_ft.loc[test_ft["대분류_생식품_cnt"] > 0, "평균_생식품_구매가격"] = test_ft["총_생식품_구매가격"] / test_ft["대분류_생식품_cnt"]
test_ft.loc[test_ft["대분류_생식품_cnt"] <= 0, "평균_생식품_구매가격"] = 0
```

```python
# 기념일,명절 구매
agg_list=[
        ('어린이날전후구매', lambda x:np.sum((x.dt.month.isin([5]))&(x.dt.day.isin(range(3,9))))),
        ('어린이날구매', lambda x:np.sum((x.dt.month.isin([5]))&(x.dt.day.isin([5])))),
        ('크리스마스전후구매', lambda x:np.sum((x.dt.month.isin([12]))&(x.dt.day.isin(range(19,27))))),
        ('크리스마스구매', lambda x:np.sum((x.dt.month.isin([12]))&(x.dt.day.isin([24,25])))),
        ('설날전후구매', lambda x: np.sum(((x.dt.month == 1) & (x.dt.day >= 20)) | ((x.dt.month == 2) & (x.dt.day <= 20)))),
        ('추석전후구매', lambda x: np.sum(((x.dt.month == 9) & (x.dt.day >= 8)) | ((x.dt.month == 10) & (x.dt.day <= 8))))

         ]

tmp = train_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
tmp = test_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
```

```python
# 지점별 방문횟수
train_tmp = pd.pivot_table(train_tr,index="ID",columns="지점코드",values="구매일시",aggfunc="nunique",fill_value=0).add_prefix("pivot_방문횟수_").reset_index()
train_ft = train_ft.merge(train_tmp,how="left",on="ID")
test_tmp = pd.pivot_table(test_tr,index="ID",columns="지점코드",values="구매일시",aggfunc="nunique",fill_value=0).add_prefix("pivot_방문횟수_").reset_index()
for col in train_tmp.columns:
    if col not in test_tmp.columns:
        test_tmp[col] = 0
test_tmp = test_tmp[train_tmp.columns]
test_ft = test_ft.merge(test_tmp,how="left",on="ID")
```

```python
# 지점별 방문비율
arr = ['pivot_방문횟수_A112000','pivot_방문횟수_A144000','pivot_방문횟수_A202000','pivot_방문횟수_A373000']
for col in arr:
    train_ft[col+"비율"] = train_ft[col]/train_ft['구매횟수']
    test_ft[col+"비율"] = test_ft[col]/test_ft['구매횟수']
```
