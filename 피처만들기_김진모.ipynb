## 결측치 처리

```python
train_ft.isnull().sum()
```

```python
test_ft.isnull().sum()
```

```python
train_ft = train_ft.fillna(0)
test_ft = test_ft.fillna(0)

train_ft.isnull().sum().sum(), test_ft.isnull().sum().sum()
```

## Feature Encoding

- ID 컬럼 제거

```python
train_ft1 = train_ft.drop(columns="ID")
test_ft1 = test_ft.drop(columns="ID")
```

```python
cols = train_ft1.select_dtypes("object").columns.tolist()
cols
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown="ignore")
```

```python
tmp = pd.DataFrame(
    enc.fit_transform(train_ft1[cols]).toarray(),
    columns =  enc.get_feature_names_out()
)

train_ft1 = pd.concat([train_ft1,tmp],axis=1).drop(columns=cols)
train_ft1.head()
```

```python
tmp = pd.DataFrame(
    enc.transform(test_ft1[cols]).toarray(),
    columns = enc.get_feature_names_out()
)
test_ft1 = pd.concat([test_ft1,tmp],axis=1).drop(columns=cols)
test_ft1.head()
```

```python
test_ft1.columns
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(train_ft1)
scaler.fit(test_ft1)
```

```python
train_ft1[train_ft1.columns] = scaler.transform(train_ft1)
train_ft1.head()
```

```python
test_ft1[test_ft1.columns] = scaler.transform(test_ft1)
test_ft1.head()
```

## 정답 데이터

```python
target = train_target["target"]
target
```

# 데이터 분리(Data Split)
- 학습데이터와 검증데이터 분리

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(train_ft1, target, test_size=0.2, random_state=42)

x_train.shape, x_valid.shape, y_train.shape, y_valid.shape
```

# 모델 학습(Train Model)


```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42)
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_valid)
pred[:5]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_valid,pred)
```
