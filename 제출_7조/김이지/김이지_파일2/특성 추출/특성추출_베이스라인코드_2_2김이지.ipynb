# 특성 공학(Feature Engineering)

## 날짜 형식으로 변환

```python
train_tr["구매일시"] = pd.to_datetime(train_tr["구매일시"])
test_tr["구매일시"] = pd.to_datetime(test_tr["구매일시"])
```

## 새로 만든 feature와 병합할 고객ID로만 이루어진 데이터프레임 생성

```python
train_ft = train_target[["ID"]]
test_ft = submit[["ID"]]

train_ft.shape, test_ft.shape
```

## train_tr 특성생성

```python
train_tr['구매일'] = train_tr['구매일시'].dt.date
test_tr['구매일'] = test_tr['구매일시'].dt.date
train_tr.shape, test_tr.shape
```

```python
train_tr['구매월'] = train_tr['구매일시'].dt.month
test_tr['구매월'] = test_tr['구매일시'].dt.month
train_tr.shape, test_tr.shape
```

```python
train_tr['구매요일'] = train_tr['구매일시'].dt.weekday
test_tr['구매요일'] = test_tr['구매일시'].dt.weekday
train_tr.shape, test_tr.shape
```

```python
train_tr['구매시간'] = train_tr['구매일시'].dt.hour
test_tr['구매시간'] = test_tr['구매일시'].dt.hour
train_tr.shape, test_tr.shape
```

## 구매일시를 이용한 특성생성

```python
agg_list = [
        # 컬럼명, 집계 방식
        ('구매횟수','nunique'),
        ('내점일수', lambda x: x.dt.date.nunique()),
        ('구매주기', lambda x: int( (x.max() - x.min()).days / x.dt.date.nunique()) ),

        ('봄_구매비율', lambda x: np.mean(x.dt.month.isin([3,4,5]))),
        ('여름_구매비율', lambda x: np.mean(x.dt.month.isin([6,7,8]))),
        ('가을_구매비율', lambda x: np.mean(x.dt.month.isin([9,10,11]))),
        ('겨울_구매비율', lambda x: np.mean(x.dt.month.isin([1,2,12]))),
        ('주구매요일', lambda x: x.dt.weekday.mode()[0]),

        #추가 내용
        ('12시이전구매비율', lambda x: np.mean(x.dt.hour <= 12)),
        ('12시이후_18시이전구매비율', lambda x: np.mean((x.dt.hour > 12) & (x.dt.hour < 18))),
        ('18시이후구매비율', lambda x: np.mean(x.dt.hour >= 18)),
        ('주중방문비율', lambda x : np.mean(x.dt.weekday < 5)),
        ('주말방문비율', lambda x : np.mean(x.dt.weekday >= 5)),
        ('평일_18시_구매비율', lambda x: ((x.dt.weekday < 5) & (x.dt.hour >= 18)).mean()),
        ('방문평균거래평균횟수', lambda x: x.nunique()/x.dt.date.nunique()),

        ('금토일방문비율', lambda x : np.mean(x.dt.weekday >= 4)),
        ('금오후토일방문비율', lambda x: np.mean((x.dt.weekday >= 5) | ((x.dt.weekday == 4) & (x.dt.hour >= 18)))),
        #('월화수방문비율', lambda x : np.mean(x.dt.weekday < 3)),

        ('주구매시간', lambda x: x.dt.hour.mode()[0] if len(x.dt.hour.mode()) > 0 else -1),



    ]

tmp = train_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

### 하루 구매 시간간격, 방문일자

```python
하루_구매_시간_간격_train = (train_tr.groupby(['ID', '구매일'])['구매일시'].max() - train_tr.groupby(['ID', '구매일'])['구매일시'].min()).dt.total_seconds()
하루_구매_시간_간격_train = 하루_구매_시간_간격_train.reset_index().rename(columns = {'구매일시' : '하루 구매 시간 간격'})

하루_구매_시간_간격_test = (test_tr.groupby(['ID', '구매일'])['구매일시'].max() - test_tr.groupby(['ID', '구매일'])['구매일시'].min()).dt.total_seconds()
하루_구매_시간_간격_test = 하루_구매_시간_간격_test.reset_index().rename(columns = {'구매일시' : '하루 구매 시간 간격'})

하루_구매_시간_간격_train.shape, 하루_구매_시간_간격_test.shape
```

```python
하루_구매_시간_간격_평균_train = 하루_구매_시간_간격_train.groupby("ID")["하루 구매 시간 간격"].mean()
하루_구매_시간_간격_평균_test = 하루_구매_시간_간격_test.groupby("ID")["하루 구매 시간 간격"].mean()

하루_구매_시간_간격_평균_train.shape, 하루_구매_시간_간격_평균_test.shape
```

```python
train_ft = pd.merge(train_ft, 하루_구매_시간_간격_평균_train, on="ID", how = "left")
test_ft = pd.merge(test_ft, 하루_구매_시간_간격_평균_test, on="ID", how = "left")

train_ft.shape, test_ft.shape
```

## 지점을 이용한 특성생성

```python
agg_list = [
          ("방문지점수","nunique"),
          ('주구매지점', lambda x: x.mode()[0]),
    ]

tmp = train_tr.groupby('ID')["지점코드"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["지점코드"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

## 브랜드코드를 이용한 특성생성

```python
agg_list = [
             ('브랜드코드_nunique', 'nunique'),
             ('주구매_브랜드', lambda x: x.mode()[0]),
             ]

tmp = train_tr.groupby('ID')["브랜드코드"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["브랜드코드"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

```python
# 브랜드별 구매3회이상 추가
brand_counts = train_tr.groupby('ID')['브랜드코드'].value_counts()
brand_counts_filtered = brand_counts[brand_counts >= 3].reset_index(name='count')
tmp = brand_counts_filtered["ID"]
tmp = tmp.drop_duplicates()
tmp=tmp.reset_index()
tmp['3회이상count'] = 1
tmp = tmp.drop(columns=['index'])
train_ft = train_ft.merge(tmp, how='left',on="ID")

brand_counts = test_tr.groupby('ID')['브랜드코드'].value_counts()
brand_counts_filtered = brand_counts[brand_counts >= 3].reset_index(name='count')
tmp = brand_counts_filtered["ID"]
tmp = tmp.drop_duplicates()
tmp=tmp.reset_index()
tmp['3회이상count'] = 1
tmp = tmp.drop(columns=['index'])
test_ft = test_ft.merge(tmp, how='left',on="ID")
```

```python
#브랜드별 평균구매액
브랜드_평균구매액 = train_tr.groupby(['ID', '브랜드코드'])['구매가격'].mean().sort_values(ascending=False).reset_index()
브랜드_평균구매액.columns = ['ID','브랜드코드', '브랜드코드_평균구매액']
브랜드_평균구매액 = 브랜드_평균구매액.groupby('ID')['브랜드코드_평균구매액'].mean().reset_index()
train_ft = train_ft.merge(브랜드_평균구매액, on='ID', how='left')

브랜드_평균구매액 = test_tr.groupby(['ID', '브랜드코드'])['구매가격'].mean().sort_values(ascending=False).reset_index()
브랜드_평균구매액.columns = ['ID','브랜드코드', '브랜드코드_평균구매액']
브랜드_평균구매액 = 브랜드_평균구매액.groupby('ID')['브랜드코드_평균구매액'].mean().reset_index()
test_ft = test_ft.merge(브랜드_평균구매액, on='ID', how='left')

train_ft.shape, test_ft.shape
```

## 중분류를 이용한 특성생성

```python
agg_list = [
            ('중분류_nunique', 'nunique'),
            ('중분류_주구매비율', lambda x: (x == x.mode()[0]).sum() / x.count()),
            # 행사키워드
            ('행사키워드_구매횟수', lambda x: x[(x.str.contains("행사")|x.str.contains("특판"))].count()),
            ]
tmp = train_tr.groupby('ID')["중분류"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left')
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["중분류"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

```python
# 아동 관련 피처생성
agg_list =[
    ('유아관련count', lambda x: x.isin([
    "캐릭터", "아동",
    "아동단품", "아동잡화", "아동특선", "아동복",
    "영창(문화)", "문구_팬시", "문구(문화)", "문구", "문화",
    "테이프", "비디오", "레코드(문화)", "잡화(문화)", "팬시코너(문화)",
    "스포츠용품", "슈즈", "완구(문화)",
    ]).sum())
]
tmp = train_tr.groupby("ID")["중분류"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
tmp = test_tr.groupby("ID")["중분류"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
```

## 대분류를 이용한 특성생성

```python
agg_list = [
            ('대분류_nunique', 'nunique'),
            ('대_주구매비율', lambda x: (x == x.mode()[0]).sum() / x.count()),
            #추가내용
            ('대분류_아동_cnt', lambda x: x.str.contains('아동').sum()),
            ('대분류_여성_cnt', lambda x: x.str.contains('여성').sum()),
            ('대분류_남성_cnt', lambda x: x.str.contains('남성').sum()),
            ('대분류_생식품_cnt', lambda x: x.str.contains('생식품').sum()),
            ]

tmp = train_tr.groupby('ID')["대분류"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left')
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["대분류"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

## 구매가격을 이용한 특성생성

```python
agg_list = [
        ('총구매액','sum'),
        ('구매건수', 'count'),
        ('평균구매액', "mean"),
        ('최대구매액', 'max'),
        ('최소구매액',lambda x: x[x > 0].min() ) ,
        ('환불금액',lambda x: x[x < 0].sum() ) ,
        ('환불건수', lambda x: ( x < 0 ).sum() ),
        ('구매금액표준편차',lambda x: x[x>0].std() ),

        #추가내용
        ('구매금액왜도', lambda x: x[x > 0].skew()),
        ('구매금액첨도', lambda x: x[x > 0].kurtosis()),
        ('구매금액_1q', lambda x: x[x > 0].quantile(0.25)),
        ('구매금액_2q', lambda x: x[x > 0].quantile(0.5)),
        ('구매금액_3q', lambda x: x[x > 0].quantile(0.75))


    ]

tmp = train_tr.groupby('ID')["구매가격"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left')
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["구매가격"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

### 같은가격재구매횟수, 비율

```python
purchase_counts = train_tr.groupby('ID')['구매가격'].count().reset_index()
purchase_counts.columns = ['ID', '거래횟수']
같은가격재구매 = train_tr[train_tr.duplicated(subset=['ID', '구매가격'], keep=False)]
같은가격재구매횟수 = 같은가격재구매.groupby('ID')['구매가격'].count().reset_index()
같은가격재구매횟수.columns = ['ID', '같은가격재구매횟수']
purchase_counts = purchase_counts.merge(같은가격재구매횟수, on='ID', how='left').fillna(0)
purchase_counts['반복구매비율'] = purchase_counts['같은가격재구매횟수'] / purchase_counts['거래횟수']
train_ft = train_ft.merge(purchase_counts[['ID', '같은가격재구매횟수','반복구매비율']], on='ID', how='left')
```

```python
purchase_counts = test_tr.groupby('ID')['구매가격'].count().reset_index()
purchase_counts.columns = ['ID', '거래횟수']
같은가격재구매 = test_tr[test_tr.duplicated(subset=['ID', '구매가격'], keep=False)]
같은가격재구매횟수 = 같은가격재구매.groupby('ID')['구매가격'].count().reset_index()
같은가격재구매횟수.columns = ['ID', '같은가격재구매횟수']
purchase_counts = purchase_counts.merge(같은가격재구매횟수, on='ID', how='left').fillna(0)
purchase_counts['반복구매비율'] = purchase_counts['같은가격재구매횟수'] / purchase_counts['거래횟수']
test_ft = test_ft.merge(purchase_counts[['ID', '같은가격재구매횟수', '반복구매비율']], on='ID', how='left')
```

```python
purchase_counts = train_tr.groupby('ID')['중분류'].count().reset_index()
purchase_counts.columns = ['ID', '중분류수']
같은중분류재구매 = train_tr[train_tr.duplicated(subset=['ID', '중분류'], keep=False)]
같은중분류재구매횟수 = 같은중분류재구매.groupby('ID')['중분류'].count().reset_index()
같은중분류재구매횟수.columns = ['ID', '같은중분류재구매횟수']
purchase_counts = purchase_counts.merge(같은중분류재구매횟수, on='ID', how='left').fillna(0)
purchase_counts['반복중분류비율'] = purchase_counts['같은중분류재구매횟수'] / purchase_counts['중분류수']
train_ft = train_ft.merge(purchase_counts[['ID', '같은중분류재구매횟수', '반복중분류비율']], on='ID', how='left')
```
