## 구매일시를 이용한 특성생성

```python
agg_list = [
        # 컬럼명, 집계 방식
        ('거래횟수','nunique'),
        ('내점일수', lambda x: x.dt.date.nunique()),
        ('구매주기', lambda x: int( (x.max() - x.min()).days / x.dt.date.nunique()) ),
        #계절별
        ('봄_구매비율', lambda x: np.mean(x.dt.month.isin([3,4,5]))),
        ('여름_구매비율', lambda x: np.mean(x.dt.month.isin([6,7,8]))),
        ('가을_구매비율', lambda x: np.mean(x.dt.month.isin([9,10,11]))),
        ('겨울_구매비율', lambda x: np.mean(x.dt.month.isin([1,2,12]))),

        ('주구매요일', lambda x: x.dt.weekday.mode()[0]),
        ('일별평균구매건수', lambda x:  x.count() / x.dt.date.nunique() ),
        ('거래개월수', lambda x: x.dt.date.astype(str).str[:-3].nunique() ),

        #추가 내용
        ('12시이전구매비율', lambda x: np.mean(x.dt.hour <= 12)),
        ('12시이후_18시이전구매비율', lambda x: np.mean((x.dt.hour > 12) & (x.dt.hour < 18))),
        ('18시이후구매비율', lambda x: np.mean(x.dt.hour >= 18)),
        ('평일_18시_구매비율', lambda x: ((x.dt.weekday < 5) & (x.dt.hour >= 18)).mean()),
        #시간별?
        ('주중방문비율', lambda x : np.mean(x.dt.weekday < 5)),
        ('주말방문비율', lambda x : np.mean(x.dt.weekday >= 5)),
        ('방문평균거래평균횟수', lambda x: x.nunique()/x.dt.date.nunique()),

        #추가 11-05
        ('일별최대구매횟수', lambda x: x.dt.date.value_counts().max()),
        ('일별최소구매횟수', lambda x: x.dt.date.value_counts().min()),
        ('일별평균구매횟수', lambda x: x.dt.date.value_counts().mean()),
        ('일별중앙값구매횟수', lambda x: x.dt.date.value_counts().median()),
        ('일별최빈값구매횟수', lambda x: x.dt.date.value_counts().mode()[0]),
        ('일별구매횟수표준편차', lambda x: x.dt.date.value_counts().std()),

        ('금토일방문비율', lambda x : np.mean(x.dt.weekday >= 4)),
        ('금오후토일방문비율', lambda x: np.mean((x.dt.weekday >= 5) | ((x.dt.weekday == 4) & (x.dt.hour >= 18)))),

        ('거래주수', lambda x: x.dt.isocalendar().week.nunique()),
        ('거래기간일수', lambda x: (x.max() - x.min()).days + 1),
        ('주구매시간', lambda x: x.dt.hour.mode()[0] if len(x.dt.hour.mode()) > 0 else -1),
        ('최근구매_구매주기', lambda x: (np.sort(x.dt.date.unique())[-1] - np.sort(x.dt.date.unique())[-2]).days if len(np.sort(x.dt.date.unique())) > 1 else 0),
        # 기념일,명절 구매
        ('어린이날전후구매', lambda x:np.sum((x.dt.month.isin([5]))&(x.dt.day.isin(range(3,9))))),
        ('어린이날구매', lambda x:np.sum((x.dt.month.isin([5]))&(x.dt.day.isin([5])))),
        ('크리스마스전후구매', lambda x:np.sum((x.dt.month.isin([12]))&(x.dt.day.isin(range(19,27))))),
        ('크리스마스구매', lambda x:np.sum((x.dt.month.isin([12]))&(x.dt.day.isin([24,25])))),
        ('설날전후구매', lambda x: np.sum(((x.dt.month == 2) & (x.dt.day >= 1)) | ((x.dt.month == 2) & (x.dt.day <= 10)))),
        ('추석전후구매', lambda x: np.sum(((x.dt.month == 9) & (x.dt.day >= 20)) | ((x.dt.month == 9) & (x.dt.day <= 29))))
    ]

tmp = train_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

```python
agg_list = [
    ('연월별평균구매금액', lambda x:x.mean()),
    ('연월별구매횟수', lambda x:x.count()),
    ('연월별총구매금액', lambda x:x.sum()),
    ('연월별최대구매금액', lambda x:x.max()),
    ('연월별최소구매금액', lambda x:x[x>0].min()),
]

tmp = train_tr.groupby(['ID', '연월'])['구매가격'].agg(agg_list).reset_index()

#각각을 열로 만들기 위해 pivot을 수행
avg_pivot = tmp.pivot(index='ID', columns='연월', values='연월별평균구매금액')
count_pivot = tmp.pivot(index='ID', columns='연월', values='연월별구매횟수')
sum_pivot = tmp.pivot(index='ID', columns='연월', values='연월별총구매금액')
max_pivot = tmp.pivot(index='ID', columns='연월', values='연월별최대구매금액')
min_pivot = tmp.pivot(index='ID', columns='연월', values='연월별최소구매금액')

#열 이름을 정리 (연월별 열로 표시)
avg_pivot.columns = [f'{month}월_avg' for month in avg_pivot.columns]
count_pivot.columns = [f'{month}월_count' for month in count_pivot.columns]
sum_pivot.columns = [f'{month}월_sum' for month in sum_pivot.columns]
max_pivot.columns = [f'{month}월_max' for month in max_pivot.columns]
min_pivot.columns = [f'{month}월_min' for month in min_pivot.columns]

#pivot 결과를 합치기
train_ft2 = pd.concat([avg_pivot, count_pivot, sum_pivot, max_pivot, min_pivot], axis=1).reset_index()

train_ft2 = train_ft2.fillna(0)

train_ft = train_ft.merge(train_ft2, on='ID', how='left')

train_ft.head()
```
