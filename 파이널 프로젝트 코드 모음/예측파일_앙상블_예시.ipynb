```python
pred = np.mean(
        [
            김이지["target"].to_numpy(),
            김이지4["target"].to_numpy(),
            이희원1["target"].to_numpy(),
            김이지5["target"].to_numpy(),
            길선종1["target"].to_numpy(),
            길선종2["target"].to_numpy(),
        ],axis=0)
pred.shape
```

```python
pred = (pred >= 0.452).astype(int)
pred
```

```python
pred.mean()
```

```python
submit = 김이지[["ID"]]
submit["target"] = pred
submit
```

```python
submit.to_csv(f"{DATA_PATH}ensemble_7조_마지막_6개_3.csv",index=False)
```
