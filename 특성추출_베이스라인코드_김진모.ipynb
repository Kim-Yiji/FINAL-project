# 특성 공학(Feature Engineering)

## 날짜 형식으로 변환

```python
train_tr["구매일시"] = pd.to_datetime(train_tr["구매일시"])
test_tr["구매일시"] = pd.to_datetime(test_tr["구매일시"])
```

## 새로 만든 feature와 병합할 고객ID로만 이루어진 데이터프레임 생성

```python
train_ft = train_target[["ID"]]
test_ft = submit[["ID"]]

train_ft.shape, test_ft.shape
```

## 구매일시를 이용한 특성생성

```python
agg_list = [
        # 컬럼명, 집계 방식
        ('구매횟수','nunique'),
        ('내점일수', lambda x: x.dt.date.nunique()),
        ('구매주기', lambda x: int( (x.max() - x.min()).days / x.dt.date.nunique()) ),
        ('주말방문비율', lambda x: np.mean(x.dt.weekday>4)),
        ('봄_구매비율', lambda x: np.mean(x.dt.month.isin([3,4,5]))),
        ('여름_구매비율', lambda x: np.mean(x.dt.month.isin([6,7,8]))),
        ('가을_구매비율', lambda x: np.mean(x.dt.month.isin([9,10,11]))),
        ('겨울_구매비율', lambda x: np.mean(x.dt.month.isin([1,2,12]))),
        ('주구매요일', lambda x: x.dt.weekday.mode()[0]),
        ('일별평균구매건수', lambda x:  x.count() / x.dt.date.nunique() ),
        ('거래개월수', lambda x: x.dt.date.astype(str).str[:-3].nunique() ),

        #추가 내용
        ('12시이전구매비율', lambda x: np.mean(x.dt.hour <= 12)),
        ('12시이후_18시이전구매비율', lambda x: np.mean((x.dt.hour > 12) & (x.dt.hour < 18))),
        ('18시이후구매비율', lambda x: np.mean(x.dt.hour >= 18)),
        ('주중방문비율', lambda x : np.mean(x.dt.weekday < 5)),
        ('주말방문비율', lambda x : np.mean(x.dt.weekday >= 5)),

        #김진모 1차
        ('12시이전구매횟수', lambda x: (x.dt.hour <= 12).sum()),
        ('12시이후_18시이전구매횟수', lambda x: ((x.dt.hour > 12) & (x.dt.hour < 18)).sum()),
        ('18시이후구매횟수', lambda x: (x.dt.hour >= 18).sum()),
        ('하루결제평균횟수', lambda x: x.nunique()/x.dt.date.nunique()),

        #추가 3
        ('일별최대구매횟수', lambda x: x.dt.date.value_counts().max()),
        ('일별최소구매횟수', lambda x: x.dt.date.value_counts().min()),
        ('일별평균구매횟수', lambda x: x.dt.date.value_counts().mean()),
        ('일별중앙값구매횟수', lambda x: x.dt.date.value_counts().median()),
        ('일변최빈값구매횟수', lambda x: x.dt.date.value_counts().mode()[0]),
        ('일별구매횟수표준편차', lambda x: x.dt.date.value_counts().std()),

    ]

tmp = train_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

```python
#추가내용
train_ft['많은_구매주기'] = train_ft['구매주기'] > train_ft['구매주기'].median()
train_ft['많은_거래개월수'] = train_ft['거래개월수'] > train_ft['거래개월수'].median()
train_ft['구매주기_거래개월'] = train_ft['많은_거래개월수'] & train_ft['많은_구매주기']

test_ft['많은_구매주기'] = train_ft['구매주기'] > train_ft['구매주기'].median()
test_ft['많은_거래개월수'] = train_ft['거래개월수'] > train_ft['거래개월수'].median()
test_ft['구매주기_거래개월'] = train_ft['많은_거래개월수'] & train_ft['많은_구매주기']
```

## 지점을 이용한 특성생성

```python
agg_list = [
          ("방문지점수","nunique"),
          ('주구매지점', lambda x: x.mode()[0]),

          #김진모
          ('최소구매지점', lambda x: x.value_counts().idxmin())
    ]

tmp = train_tr.groupby('ID')["지점코드"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["지점코드"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

## 브랜드코드를 이용한 특성생성

```python
agg_list = [
             ('브랜드코드_nunique', 'nunique'),
             ('주구매_브랜드', lambda x: x.mode()[0]),
             ('브랜드_구매_최대', lambda x: x.value_counts().max()),
             ('브랜드_구매_최소', lambda x: x.value_counts().min()),
             ('브랜드_구매_평균', lambda x: x.value_counts().mean())
             ]

tmp = train_tr.groupby('ID')["브랜드코드"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["브랜드코드"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

## 중분류를 이용한 특성생성

```python
agg_list = [
            ('중분류_nunique', 'nunique'),
            ('주구매_중분류', lambda x: x.mode()[0]),
            #추가내용
            ('중분류_아동_cnt', lambda x: x.str.contains('아동').sum()),
            #김진모
            ('중_주구매비율', lambda x: (x == x.mode()[0]).sum() / x.count())
            ]
tmp = train_tr.groupby('ID')["중분류"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left')
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["중분류"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

## 대분류를 이용한 특성생성

```python
agg_list = [
            ('대분류_nunique', 'nunique'),
            ('주구매_대분류', lambda x: x.mode()[0]),
            #추가내용
            ('대분류_아동_cnt', lambda x: x.str.contains('아동').sum()),
            #김진모
            ('_주구매비율', lambda x: (x == x.mode()[0]).sum() / x.count())
            ]

tmp = train_tr.groupby('ID')["대분류"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left')
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["대분류"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

## 구매가격을 이용한 특성생성

```python
agg_list = [
        ('총구매액','sum'),
        ('구매건수', 'count'),
        ('평균구매액', "mean"),
        ('최대구매액', 'max'),
        ('최소구매액',lambda x: x[x > 0].min() ) ,
        ('환불금액',lambda x: x[x < 0].sum() ) ,
        ('환불건수', lambda x: ( x < 0 ).sum() ),
        ('구매금액표준편차',lambda x: x[x>0].std() ),

        #추가내용
        ('구매금액왜도', lambda x: x[x > 0].skew()),
        ('구매금액첨도', lambda x: x[x > 0].kurtosis()),
        ('구매금액_1q', lambda x: x[x > 0].quantile(0.25)),
        ('구매금액_2q', lambda x: x[x > 0].quantile(0.5)),
        ('구매금액_3q', lambda x: x[x > 0].quantile(0.75))


    ]

tmp = train_tr.groupby('ID')["구매가격"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left')
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["구매가격"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

# 새 섹션

```python
train_tr['주차'] = train_tr['구매일시'].dt.isocalendar().week
train_tr['주차'] = train_tr['주차'].astype(train_tr['주차'].dtype)
test_tr['주차'] = test_tr['구매일시'].dt.isocalendar().week
test_tr['주차'] = test_tr['주차'].astype(test_tr['주차'].dtype)
```

```python
train_tr['연월'] = train_tr['구매일시'].dt.to_period('M').astype(str)
test_tr['연월'] = test_tr['구매일시'].dt.to_period('M').astype(str)
```

```python
agg_list = [
    ('연월별평균구매금액', lambda x:x.mean()),
    ('연월별구매횟수', lambda x:x.count()),
    ('연월별총구매금액', lambda x:x.sum()),
    ('연월별최대구매금액', lambda x:x.max()),
    ('연월별최소구매금액', lambda x:x[x>0].min()),
]

tmp = train_tr.groupby(['ID', '연월'])['구매가격'].agg(agg_list).reset_index()
tmp2 = test_tr.groupby(['ID', '연월'])['구매가격'].agg(agg_list).reset_index()
#각각을 열로 만들기 위해 pivot을 수행
avg_pivot = tmp.pivot(index='ID', columns='연월', values='연월별평균구매금액')
count_pivot = tmp.pivot(index='ID', columns='연월', values='연월별구매횟수')
sum_pivot = tmp.pivot(index='ID', columns='연월', values='연월별총구매금액')
max_pivot = tmp.pivot(index='ID', columns='연월', values='연월별최대구매금액')
min_pivot = tmp.pivot(index='ID', columns='연월', values='연월별최소구매금액')

avg_pivot2 = tmp2.pivot(index='ID', columns='연월', values='연월별평균구매금액')
count_pivot2 = tmp2.pivot(index='ID', columns='연월', values='연월별구매횟수')
sum_pivot2 = tmp2.pivot(index='ID', columns='연월', values='연월별총구매금액')
max_pivot2 = tmp2.pivot(index='ID', columns='연월', values='연월별최대구매금액')
min_pivot2 = tmp2.pivot(index='ID', columns='연월', values='연월별최소구매금액')


#열 이름을 정리 (연월별 열로 표시)
avg_pivot.columns = [f'{month}월_avg' for month in avg_pivot.columns]
count_pivot.columns = [f'{month}월_count' for month in count_pivot.columns]
sum_pivot.columns = [f'{month}월_sum' for month in sum_pivot.columns]
max_pivot.columns = [f'{month}월_max' for month in max_pivot.columns]
min_pivot.columns = [f'{month}월_min' for month in min_pivot.columns]

avg_pivot2.columns = [f'{month}월_avg' for month in avg_pivot2.columns]
count_pivot2.columns = [f'{month}월_count' for month in count_pivot2.columns]
sum_pivot2.columns = [f'{month}월_sum' for month in sum_pivot2.columns]
max_pivot2.columns = [f'{month}월_max' for month in max_pivot2.columns]
min_pivot2.columns = [f'{month}월_min' for month in min_pivot2.columns]

#pivot 결과를 합치기
train_ft2 = pd.concat([avg_pivot, count_pivot, sum_pivot, max_pivot, min_pivot], axis=1).reset_index()
test_ft2 = pd.concat([avg_pivot2, count_pivot2, sum_pivot2, max_pivot2, min_pivot2], axis=1).reset_index()

train_ft2 = train_ft2.fillna(0)
test_ft2 = test_ft2.fillna(0)

train_ft = train_ft.merge(train_ft2, on='ID', how='left')
test_ft = test_ft.merge(test_ft2, on='ID', how='left')

train_ft.head()
```
