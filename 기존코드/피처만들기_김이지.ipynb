# 특성 공학(Feature Engineering)

## 대대분류 생성(x)

```python
categories = {
    '가정용품': [
        '가정용품',
        '가정용품파트'
    ],
    '아동': [
        '아동',
        '아동_스포츠',
        '아동문화',
        '케주얼_구두_아동'
    ],
    '여성': [
        '여성의류파트',
        '여성정장',
        '여성캐주얼',
        '여성캐쥬얼'
    ],
    '남성': [
        '남성의류',
        '남성정장스포츠'
    ],
    '로얄': [
        '로얄부띠끄',
        '로얄부틱'
    ],
    '생식품': [
        '생식품',
        '생식품파트'
    ],
    '공산품': [
        '공산품',
        '공산품파트'
    ],
    '캐주얼': [
        '골프_유니캐쥬얼',
        '스포츠캐쥬얼'
    ],
    '영': [
        '영라이브',
        '영어덜트캐쥬얼',
        '영캐릭터',
        '영플라자'
    ],
    '잡화': [
        '잡화',
        '잡화파트'
    ],
    '옷_잡화':[
        '명품잡화',
        '패션잡화',
        '영플라자'
    ],
    '기타': []
}

# 대대분류 컬럼 생성
def classify(row):
    for category, items in categories.items():
        if row in items:
            return category
    return '기타'

train_tr['대대분류'] = train_tr['대분류'].apply(classify)
train_tr.head()
```

## 추가 사항

## 방문 시간 특성 추가

```python
train_tr['방문일자'] = train_tr['구매일시'].dt.date
train_tr
```

```python
하루_구매_시간_간격 = train_tr.groupby(['ID', '방문일자'])['구매일시'].max() - train_tr.groupby(['ID', '방문일자'])['구매일시'].min()
하루_구매_시간_간격 = 하루_구매_시간_간격.reset_index().rename(columns = {'구매일시' : '하루 구매 시간 간격'})
하루_구매_시간_간격
```

```python
하루_구매_시간_간격_평균 = 하루_구매_시간_간격.groupby("ID")["하루 구매 시간 간격"].mean()
하루_구매_시간_간격_평균
```

```python
train_tr = pd.merge(train_tr, 하루_구매_시간_간격_평균, on="ID", how = "left")
```

```python
test_tr = pd.merge(test_tr, 하루_구매_시간_간격_평균, on="ID", how = "left")
```

```python
agg_list = [
        # 컬럼명, 집계 방식
        ('하루_구매시간_간격_평균', lambda x : x)
    ]

tmp = train_tr.groupby('ID')[""].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

## 결측치 처리

```python
train_ft.isnull().sum()
```

```python
test_ft.isnull().sum()
```

```python
train_ft = train_ft.fillna(0)
test_ft = test_ft.fillna(0)

train_ft.isnull().sum().sum(), test_ft.isnull().sum().sum()
```

## Feature Encoding

- ID 컬럼 제거

```python
train_ft = train_ft.drop(columns="ID")
test_ft = test_ft.drop(columns="ID")
```

```python
cols = train_ft.select_dtypes("object").columns.tolist()
cols
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown="ignore")
```

```python
tmp = pd.DataFrame(
    enc.fit_transform(train_ft[cols]).toarray(),
    columns =  enc.get_feature_names_out()
)

train_ft = pd.concat([train_ft,tmp],axis=1).drop(columns=cols)
train_ft.head()
```

```python
tmp = pd.DataFrame(
    enc.transform(test_ft[cols]).toarray(),#fit 안됨
    columns = enc.get_feature_names_out()
)
test_ft = pd.concat([test_ft,tmp],axis=1).drop(columns=cols)
test_ft.head()
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

train_ft[train_ft.columns] = scaler.fit_transform(train_ft)
train_ft.head()
```

```python
test_ft[test_ft.columns] = scaler.transform(test_ft) #여기서 fit 절대 안됨
test_ft.head()
```

# 데이터 분리(Data Split)
- 학습데이터와 검증데이터 분리

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(train_ft, target, test_size=0.2, random_state=42)

x_train.shape, x_valid.shape, y_train.shape, y_valid.shape
```

# 모델 학습(Train Model)


```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42)
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_valid)
pred[:5]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_valid,pred)
```
